```python
print('The country with the lowest AveragePrice is ','("',(AVG_REG.min()[0]),'")','with value','("',(AVG_REG.min()[1]),'")')
print('The country with the highest AveragePrice is ','("',(AVG_REG.max()[0]),'")','with value','("',(AVG_REG.max()[1]),'")')
```
